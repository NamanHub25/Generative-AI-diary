```python
## Basic syntax rules in Python 
## Python is a case sensitive language

name = "Naman" ## Although these both carries the same meaning of the word name but still python treats them separatly due to its case sensitivity
Name = "Mehrotra"

print(name)
print(Name)
```

```python
## Indentation
## Python uses indentation to define blocks of code. Consistent uses of spaces (commonly 4) or a tab is required.

age = 32
if age > 30:
    print(age) ## After this colon the cursor on the next line started with some spaces left so this spaces is actually blocks for python

print(age) ## This age is the value of variable age declared in the very first line of this cell and is outside the if block
```

## Difference between Semantics and Syntax in Python 
Syntax is about how you write the code — the correct structure and grammar.
If syntax is wrong → Python won’t even run the program.

Semantics is about what the code means — whether the logic is correct.
If semantics is wrong → Program runs but gives wrong output.




```python
## Syntax example 

if x > 5:
    print("Hello") ## Here if we remove a colon present after 5 or a bracket after "Hello" then it would be a syntax error and if syntax is wrong the code won't execute
```

```python
## Semantics example
## Want to add two numbers
a = 10
b = 5
print(a - b) ## Here although everything is correct syntactically but the problem meant to add numbers but the logic written here if for subtracting which is a semnatics error
```

```python
## Type checking and Inferencing
type(a)

variable = 10
print(type(variable))

variable = "Naman"
print(type(variable))
```
