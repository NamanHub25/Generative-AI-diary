#### Arithmetic Operators 

```python
## Arithmetic Operations

a = 10
b = 5

add = a + b
sub = a - b
prod = a * b
quot = a / b
floor_div = a // b # Floor division --> Removes the decimal and returns the floor value
modulus_ops = a % b # Modulus operation --> Gives us the remainder
exponential_ops = a**b # a to the power b

print(add)
print(sub)
print(prod)
print(quot)
print(floor_div)
print(modulus_ops)
print(exponential_ops)
```

#### Comparison Operators 


```python
## Comparison operator
# == --> Equal to 

a = 5
b = 5

a == b

# != --> Not Equal to

str1 = "Naman"
str2 = "naman"

str1 != str2

# Greater(>) than operator 

num1 = 25
num2 = 45

num1 > num2
```

```python
## less(<) than operator

num1 = 25
num2 = 45

num1 < num2
```

```python
# Greater(>) than or equal to operator 

num1 = 45
num2 = 45

num1 >= num2
```

```python
## Less (<) than or equal to operator 

num1 = 45
num2 = 45

num1 <= num2
```

#### Logical Operators 

```python
## AND, OR and NOT operators 

a = True 
b = True

result = a and b
print(result)
```

```python
## OR operator
# Only in the case when both the variables are false then only it will give the output as false otherwise all the time it gives the value as true

X = True # If this value of X also becomes False then the final output will be False
Y = False

result = X or Y
print(result)
```

```python
## NOT Operator

X = True
not X # Gives us the reverse of a boolean variable
```

```python
## Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulus = num1 % num2
exponentiation = num1 ** num2


print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Floor Division:", floor_division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation)
```
