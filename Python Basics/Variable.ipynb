```python
## Declaring and Assiging variable

name = "Naman"
age = 22
height = 6.1

print("My name is:",name)
print("I am ", age, " years old and...")
print("my height is: ",height)
```

```python
## Naming Convention
## Variable name should be descriptive and should always start with a letter or an '_' 
## And they can contain numbers, letters and underscores in between and they are case  sensitive also.

## Valid variable names 
first_name = "Naman"
last_name = "Mehrotra"

```

```python
## Invalid variable names 

# 2age = 32
# first-name = "Naman"
# @name = "Naman"
```

```python
## Understanding variable types
# Python is dynamically typed i.e. type of variable is assigned during runtime 

age = 22 #int
name = "Naman" #str
height = 6.1 #float
is_student = True #bool

print(type(is_student))
```

```python
## Type checking and Conversion

type(height) # height has a float type

age = 32 # age has a int type
type(age)

age_str = str(age) # Now we have converted age to str type
type(age_str)
```

```python
## Some examples which we can't convert using type conversion

# name = "Naman"
# int(name) We can't convert an integer to a string

# print(type(name))
```

```python
## Dynamic Typing 
# Python allows the type of a variable to change as the program executes

var = 10
print(var, type(var))

var = "Hello"
print(var, type(var))

var = 10.12
print(var, type(var))
```

```python
## User Input

age = input("Naman enter your age: ") # This input by default is of string type and we can typecast it back to integer

print(age)
```

```python
## Simple Calculator 

first_number = float(input("Enter the first number: "))
second_number = float(input("Enter the second number: "))

sum = first_number + second_number
difference = first_number - second_number
product = (first_number*second_number)
quotient = (first_number/second_number)

print("Sum is: ", sum)
print("Difference is: ", difference)
print("Product is: ", product)
print("Quotient is: ", quotient)
```
