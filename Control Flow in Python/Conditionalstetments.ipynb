```python
## if statement
age = 20

if age >= 18:
    print("You are allowed to vote in the elections.")
```

```python
## else statement --> It is a block of code which executes if the conditional statement in the if block is False

age = 16

if age >= 18:
    print("You are eleigible to vote")
else:
    print("You are a minor")
```

```python
## elif statement --> Allows us to check for multiple conditions and it stands for "else if"

age = 20

if age < 13:
    print("You are a child")
elif age < 18: 
    print("You are a teenager")
else:
    print("You are an adult")
```

```python
## Practice Questions on conditional statements 

# Determine whether a year is a leap or not 

year = int(input("Enter the year: "))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, " is a leap year")
        else:
            print(year, " is not a leap year")
    else:
        print(year, " is a leap year")
else:
    print(year, " is not a leap year")
```

```python
## Assignments 

# Write a python script to create a simple calculator

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

operation = input("Enter the operands(+, -, *, /): ")

if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        print("Zero division error")
else:
    result = "Invalid operation"
    
print("Final result is: ", result)
```
